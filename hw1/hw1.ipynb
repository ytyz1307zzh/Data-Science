# 1 Incremental sample mean and variance

```python
def f(mu, n, new_x): # Please complete this function
    mu = n * mu + new_x
    return mu / (n + 1)
```

Function $f$ is not allowed to be used or duplicated in $g$, and $\mu'$ is not allowed to be used in $g$. Actually, it will make your $g$ function look simpler if avoid using $f$ or $\mu'$.

```python
def g(v, mu, n, new_x): # Please complete this function
    v = (n - 1) * v / n
    v += (new_x - mu) ** 2 / (n + 1)
    return v
```

```python
N = 1000000 # population size
n = 9 # sample size (old)
mu = 10 # old mean value
v = 18 # old variance
new_x = 20 # a new sample

new_mu = f(mu, n, new_x) # new mean value
new_v = g(v, mu, n, new_x) # new variance

print("new_mu", new_mu)
print("new_v", new_v)
```

# 2 Correlation analysis

```python
import pandas as pd
df = pd.read_csv("./dataset-salary/data-faculty.csv")
df
```

### 2.1 Describe the mean value, median, Q1, Q3, and variance of the feature “Count” (i.e., score of CS ranking).

```python
import numpy as np

count_array = np.squeeze(df[["Count"]].values)
count_array = np.sort(count_array)

n = len(count_array)
count_sum = 0.0

for i in count_array:
    count_sum += i
        
mean = count_sum / n
print("Mean: ", mean)

if n % 2 == 0:
    median = (count_array[n // 2] + count_array[n // 2 + 1]) / 2
else:
    median = count_array[(n + 1) // 2]
print("Median: ", median)

square_sum = 0.0
for i in count_array:
    square_sum += (i - mean) ** 2
variance = square_sum / n
print("Variance: ", variance)

q1 = count_array[n // 4]
q3 = count_array[n * 3 // 4]
print("Q1: ", q1)
print("Q3: ", q3)
```

### 2.2 Normalize the feature “Count” by MIN-MAX and print the normalized feature values.

```python
min_count = min(count_array)
max_count = max(count_array)

norm_count = (count_array - min_count) / (max_count - min_count)
norm_count
```

### 2.3 Normalize the feature “Count” by Z SCORE and print the normalized feature values.

```python
std = np.sqrt(variance)
norm_count = (count_array - mean) / std
norm_count
```

### 2.4  Calculate the correlation coefficient ρ between the original (not the normalized) “Count” and “Faculty”.

```python
fac_array = np.squeeze(df[["Faculty"]].values)
fac_array = np.sort(fac_array)

n_fac = len(fac_array)
count_sum = 0.0

for i in fac_array:
    count_sum += i
        
mean_fac = count_sum / n_fac
print("Mean Faculty: ", mean_fac)

square_sum = 0.0
for i in fac_array:
    square_sum += (i - mean_fac) ** 2
var_fac = square_sum / n_fac
print("Faculty Variance: ", var_fac)
std_fac = np.sqrt(var_fac)
print("Faculty Std: ", std_fac)

mean_count = mean
std_count = std
int_array = fac_array * count_array

n_int = n_fac
count_sum = 0.0

for i in int_array:
    count_sum += i
    
mean_int = count_sum / n_int
print("Mean Count*Faculty: ", mean_int)

cov = mean_int - mean_fac * mean_count
print("Covariance: ", cov)
corr = cov / (std_count * std_fac)
print("Correlation: ", corr)
```

### Automatically integrate the four data frames and generate a cleaned dataset that has as many rows as possible.

```python
df_faculty = pd.read_csv("./dataset-salary/data-faculty.csv")
df_graduate = pd.read_csv("./dataset-salary/data-graduate.csv")
df_salary = pd.read_csv("./dataset-salary/data-salary.csv")
df_tuition = pd.read_csv("./dataset-salary/data-tuition.csv")

merge_dict = {"Massachusetts Institute of Technology (MIT)": "Massachusetts Institute of Technology",
             "Cornell University – Ithaca, NY": "Cornell University",
              "Carnegie Mellon University (CMU)": "Carnegie Mellon University",
              "University of Washington – Main Campus": "University of Washington",
              "University of Illinois -- Urbana-Champaign": "Univ. of Illinois at Urbana-Champaign",
              "University of Texas -- Austin (Cockrell)": "University of Texas at Austin",
              "Purdue University -- West Lafayette": "Purdue University",
              "University of Michigan -- Ann Arbor": "University of Michigan",
              "University of California – San Diego (UCSD)": "University of California - San Diego"
             }
df_salary = df_salary.replace(merge_dict)
df_graduate = df_graduate.replace(merge_dict)
df_tuition = df_tuition.replace(merge_dict)

df_faculty.rename(columns={"Institution": "University"}, inplace=True)

df_faculty["University"] = df_faculty["University"].map(lambda x: x[2:])

dataset = df_graduate.join(df_salary.set_index("University"), on="University", how="outer")
dataset = dataset.join(df_tuition.set_index("University"), on="University", how="outer")
dataset = df_faculty.join(dataset.set_index("University"), on="University", how="right")
dataset
```
